### 中: prj00_例 图片新闻_Bing 找另外的图片 找另外的工作 且 抽取出 图片丶图说等信息

```python
from requests_html import HTMLSession

session = HTMLSession()

r = session.get("https://cn.bing.com/images/trending")

# 通过xpath找到工作标签
items = r.html.xpath('//img[@class="ilp_lu"]/@src')
items1=r.html.xpath('//div[@class="ilp_lf ilp_fl"]/text()')
for url in items:
    print(url)  # 获得图片src url
    display(Markdown('![]({url})'.format(url=url)))  # 展示图片用display
for text in items1:
    print(text)
```

### 难: prj00_例 图片新闻_Bing 你如何按栏目(i.e. 今日热图, 明星, 美女) 抽取出 图片丶图说等信息?

```python
import pandas as pd
url = "https://cn.bing.com/images/trending?"
session = HTMLSession()
r = session.get( url )
#今日热图
data = pd.DataFrame({
    "图片":r.html.xpath('//*[@id="ilp_t"]/div[1]/div/div/div/div/a/img/@src'),
    "图说":r.html.xpath('//*[@id="ilp_t"]/div[1]/div/div/div/div/div/a[1]/div/text()'),
})
data
```

```python
# 明显
data1= pd.DataFrame({
    "图片":r.html.xpath('//*[@id="ilp_t"]/div[2]/div/div/div/div/a/img/@src'),
    "图说":r.html.xpath('//*[@id="ilp_t"]/div[2]/div/div/div/div/div/a[1]/div/text()'),
})
data1
```

```python
#美女
data2= pd.DataFrame({
    "图片":r.html.xpath('//*[@id="ilp_t"]/div[3]/div/div/div/div/a/img/@src'),
    "图说":r.html.xpath('//*[@id="ilp_t"]/div[3]/div/div/div/div/div/a[1]/div/text()'),
})
data2
```

```python
#摄影
data3= pd.DataFrame({
    "图片":r.html.xpath('//*[@id="ilp_t"]/div[4]/div/div/div/div/a/img/@src'),
    "图说":r.html.xpath('//*[@id="ilp_t"]/div[4]/div/div/div/div/div/a[1]/div/text()'),
})
data3
```

```python
#壁纸
data4= pd.DataFrame({
    "图片":r.html.xpath('//*[@id="ilp_t"]/div[5]/div/div/div/div/a/img/@src'),
    "图说":r.html.xpath('//*[@id="ilp_t"]/div[5]/div/div/div/div/div/a[1]/div/text()'),
})
data4
```

```python
#动画
data5= pd.DataFrame({
    "图片":r.html.xpath('//*[@id="ilp_t"]/div[6]/div/div/div/div/a/img/@src'),
    "图说":r.html.xpath('//*[@id="ilp_t"]/div[6]/div/div/div/div/div/a[1]/div/text()'),
})
data5
```

```python
#合并多个图标concat，axis=1 横向的合并  axis=0 纵向的合并  
pd.concat([data,data1,data2,data3,data4,data5],axis=1,keys=["热图","明星","美女","摄影","壁纸","动画"],join='inner')
```
